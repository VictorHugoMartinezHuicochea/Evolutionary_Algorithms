```python
#Importamos librerias
import sympy
import copy
import random


#Evalua un conjunto de Vectores
def Evaluacion(Subject,SelectedFunction):
  #Donde se guardan sus evaluaciones
  Proficiency=[]
  Evaluation=0

  #Establecemos nuestras funciones
  #Funcion 1
  if(SelectedFunction==1):

    for i in range(0,len(Subject)):
      Evaluation=0
      for j in range(0,len(Subject[0])):
        Evaluation+=Subject[i][j]**2
      Proficiency.append(Evaluation)
    return Proficiency

  #Funcion 2
  if(SelectedFunction==2):
    for i in range(0,len(Subject)):
      Evaluation=sympy.sin(Subject[i][0]+Subject[i][1])+(Subject[i][0]-Subject[i][1])**2-1.5*Subject[i][0]+2.5*Subject[i][1]+1
      Proficiency.append(Evaluation)
    return Proficiency


  #Funcion 3
  if(SelectedFunction==3):

    for i in range(0,len(Subject)):
      Evaluation=0
      for j in range(0,len(Subject[0])-1):
        Evaluation+=100*(Subject[i][j+1]-Subject[i][j]**2)**2+(Subject[i][j]-1)**2
      Proficiency.append(Evaluation)
    return Proficiency





#Utilizamos Mutacion uniforme
def MutacionUniforme(Father,SelectedFunction,Generation,MaxGeneration):
  Mutated=copy.deepcopy(Father)

  #Establecemos otro volado
  volado=random.uniform(0,1)
  ModifiedElement=random.randrange(len(Mutated))

  #Establecemos dominios segun la funcion
  if(SelectedFunction==1):
    Dominio=[-5.12,5.12]
  if(SelectedFunction==2):
    if(ModifiedElement==0):
      Dominio=DominioX=[-1.5,4]
    else:
      Dominio=DominioY=[-3,4]
  if(SelectedFunction==3):
    Dominio=[-5,10]

  #Especificamos suma o resta y Mutamos
  if (volado<0.5):
    Element=Dominio[1]-Mutated[ModifiedElement]
    Mutated[ModifiedElement]=Mutated[ModifiedElement]+DeltaOperation(Generation,Element,MaxGeneration)
  else:
    Element=Mutated[ModifiedElement]-Dominio[0]
    Mutated[ModifiedElement]=Mutated[ModifiedElement]-DeltaOperation(Generation,Element,MaxGeneration)

  #Retornamos Vector
  return(Mutated)



def DeltaOperation(Generation,Element,MaxGeneration):
  r=random.uniform(0,1)
  Delta=Element*(1-(r**(1-(Generation/MaxGeneration)**5)))
  return Delta






#Vemos quien sera la nueva generacion
def Supervivencia(Childrens,ProficiencyChildren,SelectedFunction,NumberOfTotalParents):
  #Combinamos y guardamos sus aptitudes
  Proficiency=copy.deepcopy(ProficiencyChildren)
  Population=copy.deepcopy(Childrens)

  #Escogemos a los nuevos padres y matamos a los peores
  NewFathers=[]
  for i in range(0,NumberOfTotalParents):
        c=min(Proficiency)
        for j in range(0,len(Proficiency)):
            if(Proficiency[j]==c):
                NewFathers.append(Population[j])
                Proficiency[j]=100000000000000000000001
                break

  #Reevaluamos
  NewProficiencyParents=Evaluacion(NewFathers,SelectedFunction)

  return NewFathers,NewProficiencyParents







#Iniciamos Algoritmo---------------------------------------------------------------------------------------------------------------------
if __name__ == "__main__":
  #Especificamos Funcion y n-------------------------------------------------------------------------------------------------------------
  SelectedFunction=2
  n=20

  #Definimos Datos importantes
  NumberOfTotalParents=50
  Parents=[]
  Generation=0
  MaxGeneration=100


  #Definimos Dominios de las Funciones
  if(SelectedFunction==1):
    Dominio=[-5.12,5.12]
  if(SelectedFunction==2):
    DominioX=[-1.5,4]
    DominioY=[-3,4]
  if(SelectedFunction==3):
    Dominio=[-5,10]



  #Creamos Poblacion Inicial
  if(SelectedFunction==1 or SelectedFunction==3):
    for i in range(0,NumberOfTotalParents):
      BornParent=[]
      for j in range(0,n):
        BornParent.append(random.uniform(Dominio[0],Dominio[1]))
      Parents.append(BornParent)
  else:
    for i in range(0,NumberOfTotalParents):
      Parents.append([random.uniform(DominioX[0],DominioX[1]),random.uniform(DominioY[0],DominioY[1])])

  #Evaluamos Padres
  ProficiencyParents=Evaluacion(Parents,SelectedFunction)




  #Iniciamos Evolucion
  while(Generation!=MaxGeneration):

    #Buscamos a los padres para ir a la mating pool
    Childrens=[]

    #Un padre genera un hijo
    for i in range(0,2*len(Parents)):
      j=int(i/2)
      Childrens.append(MutacionUniforme(Parents[j],SelectedFunction,Generation,MaxGeneration))

    #Obtenemos la aptitud de los hijos
    ProficiencyChildren=Evaluacion(Childrens,SelectedFunction)

    #Generamos nuevos padres
    Parents,ProficiencyParents=Supervivencia(Childrens,ProficiencyChildren,SelectedFunction,NumberOfTotalParents)

    #Aumentamos generacion
    Generation+=1



  #Imprimimos resultados
  print("Despues de ",Generation,"Generaciones, obtuvimos a estos padres: ")
  print(Parents)
  print("Cuyo desempeño es:")
  print(ProficiencyParents,"\n\n")
  c=min(ProficiencyParents)
  for i in range(0,len(Parents)):
    if(c==ProficiencyParents[i]):
      c=i
      break
  print("Mejor Punto Alcanzado:",Parents[i], "\n Evaluado da:", min(ProficiencyParents))


  print("Fin del Programa :D")
```
