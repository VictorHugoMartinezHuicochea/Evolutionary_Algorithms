#Mutacion

```python
def Mutacion(R1,R2,R3):
  Vector=copy.deepcopy(R1)
  #Generamos a F aleatorio entre 0 y 2 (Se dejo en 0.5 para fines practicos)
  F=0.5

  #Vamos modificando al nuevo vector de prueba
  for i in range(len(R1)):
    Vector[i]=R1[i]+F*(R2[i]-R3[i])

  return Vector
```

```python
def MutacionTripleT(R1,RB,R2,R3):
  Vector=copy.deepcopy(R1)
  #Generamos a F aleatorio entre 0 y 2 (Se dejo en 0.5 para fines practicos)
  F=0.5

  #Vamos modificando al nuevo vector de prueba
  for i in range(len(R1)):
    Vector[i]=R1[i]+F*(RB[i]-R1[i])+F*(R2[i]-R3[i])

  return Vector
```

#Supervivencia

```python
def Supervivencia(Parents,Pruebas,ProficiencyParents,ProficiencyPruebas):
  #Escogemos a los nuevos padres a partir de su comparacion
  NewFathers=[]

  for i in range(len(Parents)):
    if(ProficiencyParents[i]<ProficiencyPruebas[i]):
      NewFathers.append(Parents[i])
    else:
      NewFathers.append(Pruebas[i])

  #Reevaluamos
  NewProficiencyParents=Evaluacion(NewFathers)

  return NewFathers,NewProficiencyParents
```

```python
def Rand(NumberOfTotalParents):
  #Generamos la cantidad de aleatorios
  Aleatorios=[]
  #Evita que se repitan
  while(len(Aleatorios)!=3):
    Selected=random.randrange(NumberOfTotalParents)
    #Se crea una lista de mutacion
    if(Selected not in Aleatorios):
      Aleatorios.append(Selected)

  return Aleatorios
```

```python
def Best(NumberOfTotalParents,ProficiencyParents):
  #Encontramos al mejor
  Best=min(ProficiencyParents)
  Ub=ProficiencyParents.index(Best)



  #Generamos la cantidad de aleatorios
  Aleatorios=[]
  Aleatorios.append(Ub)
  #Evita que se repitan
  while(len(Aleatorios)!=3):
    Selected=random.randrange(NumberOfTotalParents)
    #Se crea una lista de mutacion
    if(Selected not in Aleatorios):
      Aleatorios.append(Selected)

  return Aleatorios
```

```python
def TripleT(NumberOfTotalParents,ProficiencyParents,i):
  #Encontramos al mejor
  Best=min(ProficiencyParents)
  Ub=ProficiencyParents.index(Best)

  #Generamos la cantidad de aleatorios
  Aleatorios=[]
  Aleatorios.append(i)
  Aleatorios.append(Ub)
  #Evita que se repitan
  while(len(Aleatorios)!=4):
    Selected=random.randrange(NumberOfTotalParents)
    #Se crea una lista de mutacion
    if(Selected not in Aleatorios):
      Aleatorios.append(Selected)

  return Aleatorios
```
