```python
#Escuela Superior de Fisico y Matematicas
#Licenciatura En Matematica Algoritmica
#Fundamentos de la inteligencia Artificial
#Profesora: Lourdes Fabiola Uribe Richaud
#Alumno:
    #Victor Hugo Martinez Huicochea

#Importamos librerias
import sympy
import copy
import random

def minimo(busc):
    c=100000
    for i in range(0,len(busc)):
      if(busc[i]<c):
        c=busc[i]
    return c



def seleccion(resul,bestpad,padt,padres):
    #Buscamos a los mejores resultados
    busc=copy.deepcopy(resul)
    best=[]
    for i in range(0,bestpad):
        c=minimo(busc)
        for j in range(0,padt):
            if(busc[j]==c):
                best.append(padres[j])
                busc[j]=100001

    return best

#Definimos la funcion diferencial
def dif(t,y,Tot):
    b=5
    r=random.uniform(0,1)
    l=y*(1-(r**((1-t/Tot)**b)))
    return l

def pool(best,bestpad,Tot,t):
    hijos=[]
    #Hijo por mutacion
    elegido=random.randrange(bestpad-1)

    #Mutacion uniforme
    volado=random.uniform(0,1)

    #Copiamos al que mutaremos
    Mutacionunif1=best[elegido]
    Mutacionunif2=best[elegido]


    #Aplicamos la mutacion uniforme
    if(volado<0.5):
        Mutacionunif1[0]=Mutacionunif1[0]+dif(t,100-Mutacionunif1[0],Tot)
        Mutacionunif2[1]=Mutacionunif2[1]+dif(t,100-Mutacionunif2[1],Tot)
    else:

        Mutacionunif1[0]=Mutacionunif1[0]-dif(t,Mutacionunif1[0]+100,Tot)
        Mutacionunif2[1]=Mutacionunif2[1]-dif(t,Mutacionunif2[1]+100,Tot)
    hijos.append(Mutacionunif1)
    hijos.append(Mutacionunif2)


    return hijos


def batalla(padres,hijos,func,resul,padt):
    busca=copy.deepcopy(resul)
    #Agregamos resultados de hijos
    for i in range(0,len(hijos)):
        eva=func.subs([(x,hijos[i][0]),(y,hijos[i][1])])
        busca.append(eva)
        padres.append(hijos[i])
    #Escogemos a los nuevos padres y matamos a los peores
    newpad=[]
    for i in range(0,padt):
        c=minimo(busca)
        for j in range(0,len(busca)):
            if(busca[j]==c):
                newpad.append(padres[j])
                busca[j]=100001
                break


    #Reevaluamos
    resul=[]
    for i in range(0,padt):
        eva=func.subs([(x,newpad[i][0]),(y,newpad[i][1])])
        resul.append(eva)
    return newpad,resul

if __name__ == "__main__":

    #Creamos padres
    Tot=20000
    padt=6
    bestpad=4
    padres=[]
    gen=0
    for i in range(0,padt):
        padres.append([random.uniform(-100, 100),random.uniform(-100, 100)])

    meta=False

    #Establecemos la funcion a evaluar
    x=sympy.symbols("x")
    y=sympy.symbols("y")

    func=x**2+y**2
    #func=100*(y-x**2)**2+(1-x)**2

    #Guardamos sus evaluaciones
    resul=[]
    for i in range(0,padt):
        eva=func.subs([(x,padres[i][0]),(y,padres[i][1])])
        if(eva==0):
            print("Fin, sanseacabo :/")
            meta=True
        resul.append(eva)


    while(meta!=True):
        best=seleccion(resul,bestpad,padt,padres)
        hijos=pool(best,bestpad,Tot,gen)
        padres,resul=batalla(padres,hijos,func,resul,padt)

        for i in range(0,len(resul)):
            if(resul[i]==0):
                meta=True
                direc=i

        if(gen==Tot):
            min=100000
            for i in range(0,len(resul)):
                if(resul[i]<min):
                  min=resul[i]
                  direc=i
            break;

        gen+=1


    print("Para obtener el minimo x e y deben ser: ", padres[direc])
    print("Tomo un total de", gen, " generaciones")
    print("Fin del programa :D")
```
